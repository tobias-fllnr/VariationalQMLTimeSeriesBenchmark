```python
import reservoirpy.datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-deep")
plt.rcParams.update({
    "text.usetex": True,  # Use LaTeX for text rendering
    "font.family": "serif",  # Use serif font
    "font.serif": ["Computer Modern Roman"],  # Use default LaTeX font
})
```

Generate Henon Data

```python
time_steps = 10000
data = reservoirpy.datasets.henon_map(time_steps)
df = pd.DataFrame({'x_coords': data[:,0], 'y_coords': data[:,1]})
df_data = df.loc[:999, :] 
df_data.to_csv(f'henon_1000.csv', index=False)
```

Plot Henon Data

```python
data_henon = np.array(df_data)

x = data_henon[:,0]
y = data_henon[:,1]

# Create the figure and axis
fig = plt.figure(figsize=(3.5, 3))
ax = fig.add_subplot(111)  # No 'projection' needed for 2D plots

# Plot the data as a line
ax.plot(x, y, linestyle='-')

# Set labels and title
ax.set_xlabel('$x(t)$')
ax.set_ylabel('$y(t)$')
fig.suptitle('Henon Attractor')

# Save the plot
plt.tight_layout()
plt.savefig("./Plots/henon_attractor.pdf")
```

```python
fig, axes = plt.subplots(2, 1, figsize=(2.5, 2.5))

# Plot x(t)
axes[0].plot(x)
axes[0].set_ylabel('$x(t_{step})$')

# Plot y(t)
axes[1].plot(y)
axes[1].set_ylabel('$y(t_{step})$')

# Set common x-axis label
axes[1].set_xlabel('$t_{step}$')
#fig.suptitle('HÃ©non Map Time Series Data')
# Adjust spacing between subplots
plt.tight_layout()

plt.savefig("./Plots/henon_timeseries.pdf")
```
