```python
import reservoirpy.datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-deep")
plt.rcParams.update({
    "text.usetex": True,  # Use LaTeX for text rendering
    "font.family": "serif",  # Use serif font
    "font.serif": ["Computer Modern Roman"],  # Use default LaTeX font
})
```

Generate Lorenz Data

```python
time_steps = 10000
data = reservoirpy.datasets.lorenz(time_steps)
df = pd.DataFrame({'x_coords': data[:,0], 'y_coords': data[:,1], 'z_coords': data[:,2]})
df_data = df.loc[500:1499, :] # leave out first 500 time steps
df_data.to_csv(f'lorenz_1000.csv', index=False)
```

Plot Lorenz Data

```python
data_lorenz = np.array(df_data)

x = data_lorenz[:,0]
y = data_lorenz[:,1]
z = data_lorenz[:,2]
fig = plt.figure(figsize=(3.5, 3))
ax = fig.add_subplot(111, projection='3d')

# Plot the data as a line
ax.plot(x, y, z, linestyle='-')

# Set labels (optional)
ax.set_xlabel('$x(t)$')
ax.set_ylabel('$y(t)$')
ax.set_zlabel('$z(t)$')
fig.suptitle('Lorenz Attractor')
# plt.tight_layout()
# plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)

# ax.view_init(elev=20, azim=-75) 
plt.savefig("./Plots/lorenz_attractor.pdf")
```

```python
fig, axes = plt.subplots(3, 1, figsize=(2.5, 2.5))

# Plot x(t)
axes[0].plot(x)
axes[0].set_ylabel('$x(t_{step})$')

# Plot y(t)
axes[1].plot(y)
axes[1].set_ylabel('$y(t_{step})$')

# Plot z(t)
axes[2].plot(z)
axes[2].set_ylabel('$z(t_{step})$')

# Set common x-axis label
axes[2].set_xlabel('$t_{step}$')
#fig.suptitle('Lorenz Attractor Time Series Data')
# Adjust spacing between subplots
plt.tight_layout()

plt.savefig("./Plots/lorenz_timeseries.pdf")
```
